```python
def count_video_text_length_number_of_words(video_input_text):
    return len(video_input_text.split())
```

```python
def unique_count_video_text_length_number_of_words(video_input_text):
    return len(set(video_input_text.split()))
```

```python
def get_counts_terms_tuples_table(video_input_text):
    r = dict()
    
    for t in video_input_text.split():
        if t not in r:
            r[t] = 1
        else:
            r[t]+= 1
            
    return r
```

```python
def get_word_term_frequency_sample_probability(video_input_text):
    n = count_video_text_length_number_of_words(video_input_text)
    tab= get_counts_terms_tuples_table(video_input_text)
    r = dict()
    
    for t in tab:
        r[t] = float(tab[t])/n
    
    return r
```

```python
def clean_string(s):
    import string
    return s.translate(str.maketrans('', '', string.punctuation)).lower().replace('\n', ' ')
```

```python
import re

DOLLAR_PRICE_REGEX = '\$\d+\.?\d?\d?'

def get_word_worth_in_mean_or_average_amazon_api_dollars(word):
    t = '+'.join(word.split()) if len(word.split()) > 1 else word
    r = !curl -k -sA "Chrome" -L https://www.google.com/search?q=$t+prices&source=lnms&tbm=shop
    # (?:\d+\.)?\d+  (?:[\£\$\€]{1}[\d]+\.?\d*)
    prices = (re.findall(DOLLAR_PRICE_REGEX, ''.join([rx for rx in r])))
    prices = sorted([float(p[1:]) for p in prices])
    return ["$" + str(p) for p in prices]
```

### Average Mean Price of Word Worthy / Term Value (Online Ebay/ Google SERG Listings Dollar Costs)

```python
def is_costs_list(cost):
    return type(cost) == list or "$" in str(cost)

idxs = price_table[price_table['COST'].apply(is_costs_list)].index


for i in idxs:
    C = price_table.loc[i].COST
    
    if type(C) == list:
        costs_list = [float(c.strip()[1:]) for c in C]
    elif type(C) == str:
        costs_list = [float(C.strip()[1:])]
    price_table.set_value(i, 'MEAN', np.mean(costs_list))
    
for j in set(price_table.index) - set(idxs):
    price_table.set_value(j, 'MEAN', float(price_table.loc[j].COST))
```
